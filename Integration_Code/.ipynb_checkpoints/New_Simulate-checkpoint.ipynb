```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
import pyroomacoustics as pra
from scipy.io import wavfile
from scipy.io import savemat
from scipy.signal import correlate
import IPython
import heapq
```

```python
#uni_white_noise = np.random.uniform(-1,1,57040)
uni_white_noise = np.random.uniform(-0.01,0.01,5000)
uni_white_noise[0] = 1

plt.plot(uni_white_noise)
plt.show()
print(uni_white_noise.shape)
```

```python
def get_rir(size='medium', absorption='medium', order = 0):
    if absorption=='high':
        absor = 0.9
    elif absorption=='medium':
        absor = 0.3
    elif absorption=='low':
        absor = 0.0
    else:
        raise ValueError("The absorption parameter can only take values ['low', 'medium', 'high']")
    
    if size=='large':
        size_coef = 5.
    elif size=='medium':
        size_coef = 2.5
    elif size=='small':
        size_coef = 1.
    else:
        raise ValueError("The size parameter can only take values ['small', 'medium', 'large']")
        
        
    pol = size_coef * np.array([[0,0], [0,7], [10,7], [10,0]]).T
    print("Length and Breadth:",pol)
    room = pra.Room.from_corners(pol, fs=16000, max_order=order, absorption=absor)

    # Create the 3D room by extruding the 2D by a specific height
    room.extrude(size_coef * 4.5, absorption=absor)
    print("Height:",size_coef * 4.5)

    # Adding the source
    room.add_source(size_coef * np.array([2.25, 1.75, 0.75]), signal=uni_white_noise)
    print(size_coef * np.array([2.25, 1.75, 0.75]))

    # Adding the microphones
    R = size_coef * np.transpose(np.array([[2.75,2.75, 2.75, 2.75, 2.75, 3.,   3.,   3.,   3.,   3.],
                                           [1.25, 1.5,  1.75, 2.,   2.25, 1.25, 1.5,  1.75, 2.,   2.25],
                                           [0.75, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75, 0.75]]))
    R = np.transpose(R)
    print(R)
    room.add_microphone_array(pra.MicrophoneArray(R, room.fs))
    print(room.fs)
    
    print(dir(room.walls))
    #print(dir(room))
    
    fig, ax = room.plot()
    ax.set_xlim([0, 10])
    ax.set_ylim([0, 7])
    ax.set_zlim([0, 4.5])
    plt.show()
    
    # Compute the RIR
    room.compute_rir()

    room.simulate()
    room.mic_array.to_wav('aaa.wav', norm=True, bitdepth=np.int16)
    
    # Display the audio file
    #fs_result, result = wavfile.read('aaa.wav')
    #IPython.display.display(IPython.display.Audio(result, rate=fs_result))
    
    return room.rir
```

```python
rir_small = get_rir(size='small', absorption='low', order = 1)

rir_small = np.array(rir_small)
print(rir_small.shape)

plt.plot(rir_small[0][0])
plt.show()

plt.plot(rir_small[1][0])
plt.show()
```

```python
rir_done = []

for i in range(10):
    for row in rir_small[i]:
        rir_done.append(row)
        print(len(row))
```

```python
out = []

for i in range(10):
    o = np.convolve(uni_white_noise, rir_done[i])
    out.append(o)
```

```python
plt.plot(out[0])
plt.show()

peaks = []
thresh = 0.045

for i in range(len(out[0])):
    if(out[0][i] > thresh):
        peaks.append(i)
#print(peaks)
        
idx = []
for i in range(len(peaks) - 1):
    if(peaks[i+1] - peaks[i] < 3):
        idx.append(i)
        
peaks_del = np.delete(peaks, idx)
        
print(np.array(peaks_del))
```
